```python
import numpy as np
import numpy.random as random
import scipy as sp
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
%matplotlib inline
%precision 3

#指定された数の乱数を生成
N=10000
x=np.random.uniform(0.0,1.0,N)
y=np.random.uniform(0.0,1.0,N)
#print(x)
#print(y)

#円周率の計算
z=(x**2+y**2)**0.5
incount=sum([i<1 for i in z])
calc_pi=4*incount/N
#print(incount)
print('円周率（モンテカルロ法）：', calc_pi)

#乱数をプロットして描画
#plt.figure()

figure, axes = plt.subplots(figsize=(10,10))
axes.plot(x,y,'o', markersize=2)
draw_circle = plt.Circle((0, 0), 1, fill=False, ec="red", lw=3)
axes.add_artist(draw_circle)
plt.show()

```
